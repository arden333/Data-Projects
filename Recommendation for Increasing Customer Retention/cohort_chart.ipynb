#코호트 분석 실습

##1.전처리

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('https://github.com/springcoil/marsmodelling/blob/master/relay-foods.xlsx?raw=true', 
                 sheet_name='Purchase Data - Full Study')

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.UserId.nunique()
```

```python
# 불필요한 컬럼 삭제
df.drop(['CommonId','PupId','PickupDate','TotalCharges'],
        axis=1,inplace=True)
df.head()
```

```python
# orderdate 연-월 형태로 변환
df['OrderPeriod']=df['OrderDate'].dt.strftime('%Y-%m')
df.head()
```

##2.코호트 기준 설정 및 분류

유저의 첫번째 주문을 토대로 유저별 코호트 그룹 설정

```python
df.set_index('UserId',inplace=True)
df.head()
```

```python
FirstOrder = df.groupby(df.index)['OrderDate'].min()
FirstOrder = FirstOrder.dt.strftime('%Y-%m')
df['FirstOrder'] = FirstOrder
df.head()
```

사용자 유지율을 파악하기 위한 코호트 분석에서는 각 방문을 모두 집계하는 것이 아니라 순방문자(UV)로 집계해야 함

```python
df.reset_index(inplace=True)
cohorts = df.groupby(['FirstOrder', 'OrderPeriod'], as_index=False)['UserId'].nunique()
cohorts.rename({'UserId':'TotalUsers'}, axis=1, inplace=True)
cohorts.head()
```

- 2009년 1월에 첫 구매한 사람중 2월에도 구매한 사람이 8명, 3월에 구매한 사람은 10명이라는 의미

image.png

코호트표에서 y축 기준 첫번째 그룹이 n번의 기간이 있다면 두번째 그룹은 n-1번, 마지막 그룹은 1번이어야 함

```python
each_period = cohorts['FirstOrder'].value_counts().sort_index()
each_period
```

- 2009-02가 14여야 하는데 13임

```python
cohorts[cohorts['FirstOrder']=='2009-02']
```

```python
cohorts[cohorts['FirstOrder']=='2009-03'].tail()
```

- 다른 데이터는 2010-03 데이터 있으나 첫구매가 2020-01인 경우는 없음

```python
# 데이터 없는 부분은 0 값으로 채움
cohorts.loc[cohorts.shape[0],'FirstOrder']='2009-02'
cohorts.loc[cohorts.shape[0]-1,'OrderPeriod']='2010-03'
cohorts.loc[cohorts.shape[0]-1,'TotalUsers']=0
cohorts.sort_values(['FirstOrder','OrderPeriod'],inplace=True)
```

```python
each_period=cohorts['FirstOrder'].value_counts().sort_index()
each_period
```

```python
# 기간별 레이블 붙이기
cohortperiod = []
for x in each_period: 
    for y in range(x):
        cohortperiod.append(y)
```

```python
cohorts['CohortPeriod']=cohortperiod
cohorts.head(20)
```

```python
cohorts.set_index(['FirstOrder','CohortPeriod'],inplace=True)
cohorts.head()
```

unstack(1)을 하면 생성된 멀티인덱스 중 두번째 인덱스가 컬럼으로 변경되고, 그 결과 y축은 코호트, x축은 기간으로 표현됨

```python
cohorts = cohorts['TotalUsers'].unstack(1)
cohorts
```

##3.코호트별 사용자 유지율

- 처음의 순 방문자 수에 비해 얼마나 유지되고 있는지 알기 위한 것이므로 각 데이터를 첫 코호트 기간의 순방문자 수로 나눔

```python
each_cohort = cohorts[0]
each_cohort
```

```python
retention_matrix = cohorts.divide(cohorts[0], axis=0)
retention_matrix
```

##4.시각화

```python
# cohort chart
plt.figure(figsize=(15,10))
plt.title('Cohort Analysis', fontsize=20)
sns.heatmap(retention_matrix, annot=True, fmt='.0%')
plt.show();
```

```python
# retention curve
retention_avg = retention_matrix.mean(axis=0) * 100

plt.figure(figsize=(10,8))
# sns.lineplot(data=retention_avg)
plt.plot(range(0, len(retention_avg)), retention_avg)
plt.title('Retention curve', fontsize=20)
plt.ylabel('percentage')
plt.show()
```
